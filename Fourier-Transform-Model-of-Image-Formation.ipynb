# Activity 1. Familiarization with discrete FFT


```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from numpy.fft import fft2, fftshift,fft
```

```python
W =mpimg.imread('whitecirc1.png')
#Conversion from RGB to Grayscale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
W_gray = rgb2gray(W)    
#Applying fft2
W_fft2 = np.abs(fft2(W_gray))
#interchanging the quadrants 
W_fshift = np.abs(fftshift(fft2(W_gray)))
#Applying fft2 again
W_fft22 = np.abs(fft2(fft2(W_gray)))

#Separating real part and complex part
W_fshift_r = np.real(W_fshift)
W_fshift_c = np.imag(W_fshift)

plt.imshow(W_gray, cmap = plt.get_cmap('gray'))
plt.title('Input Image')
plt.show()
plt.subplot(1,2,1),plt.imshow(W_fft2,'gray')
plt.title('fft2')
plt.subplot(1,2,2),plt.imshow(W_fft22,'gray')
plt.title('fft2 twice')
plt.show()
plt.imshow(W_fshift,cmap='gray')
plt.title('with fftshift')
plt.show()

plt.subplot(131),plt.imshow(W_fshift_r,cmap='gray')
plt.title('real part')
plt.subplot(132),plt.imshow(W_fshift_c,cmap='gray')
plt.title('complex part')
plt.show()
```

```python
B = np.linspace(0,90,1000)
x = np.sin(B) + 2*np.sin(3*B)
t = np.linspace(0, 0.5, 200)

fft = np.fft.fft(x)
T = t[1] - t[0] 
N = x.size

f = np.linspace(0, 1 / T, N)

plt.plot(B,x,'k')
plt.show()

plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.bar(f[:N // 2], np.abs(fft)[:N // 2] * 1 / N, width=1.5,color='k')  # 1 / N is a normalization factor
plt.show()


```
