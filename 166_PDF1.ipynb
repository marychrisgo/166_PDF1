# Activity 1: Familiarization with discrete FFT

<p style='text-align: justify;'>

The first activity deals with being familiar with Fourier Transform. The first part demonstrated an image of a while circle drawin in 128x128 pixels. The FFT (twice) of the image showed something similar to the original image but rotated by 180 degrees. For the circles, it's not that visible. But for letter 'A', it is obvious. This is different for other patterns such as sinusoid along x, double slit, square function, and 2D Gaussian bell curve. Part of our past experiments was doing the double slit and knowing its Fourier transform, and in theory, this prove what we see as light passes through the slits. 

</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from numpy.fft import fft2, fftshift,fft,ifft2
```

### 1. 1 For FFT of Circle Image

```python
I =mpimg.imread('121x128.jpg')

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
I_gray = rgb2gray(I)    

I_fft2 = np.abs(fft2(I_gray))
I_fshift = np.abs(fftshift(fft2(I_gray)))
I_fft22 = np.abs(fft2(fft2(I_gray)))

#Separating real part and complex part
I_fshift_r = np.real(I_fshift)
I_fshift_c = np.imag(I_fshift)


plt.imshow(I_fft22,'gray')
plt.title('fft2 twice')
plt.show()
plt.imshow(I_fshift,cmap='gray')
plt.title('with fftshift')
plt.show()

```

### 1.3 Sinusoid along x

```python
S = np.linspace(0,90,1000)
l = np.sin(S) + 2*np.sin(3*S)
t = np.linspace(0, 0.5, 200)
fft = np.fft.fft(l)
T = t[1] - t[0] 
N = l.size

f = np.linspace(0, 1 / T, N)

plt.plot(S,l,'k')
plt.show()

plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.bar(f[:N // 2], np.abs(fft)[:N // 2] * 1 / N, width=1.5,color='k')  # 1 / N is a normalization factor
plt.show()

```

# Activity 3: Template Matching using correlation

<p style='text-align: justify;'>

The template matching is a pattern recognition technique. This is the most used application of correlation. By getting the FT of both images, we use the conj() function. In here, we get the complex conjugate of "A" and is being applied to the phrase picture. The function will then multiply the element by element.

</p>

```python
R = mpimg.imread('rain in spain.jpg')
A = mpimg.imread('A.jpg')
#Conversion from RGB to Grayscale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
# To grayscale
R_gray = rgb2gray(R)
A_gray = rgb2gray(A)
# applying fft2
R_fft2 = np.abs(fft2(R_gray))
A_fft2 = np.abs(fft2(A_gray))
# Multiply element per element of A and Rain
M = R_fft2.conj()*A_fft2
# Inverse fft
i_M = np.abs(ifft2(M))

plt.subplot(121), plt.imshow(R_gray,cmap='gray')
plt.subplot(122), plt.imshow(A_gray,cmap='gray')
plt.show()
plt.imshow(i_M,cmap='gray')
```

# Activity 4: Edge Detection

<p style='text-align: justify;'>

Last for Activity 1 deals with the detection of the edges of the word VIP. By convolving the pattern and the image, edges were detected. For the diagonal pattern, it was somehow messed up because of the fact that diagonal pattern has both vertical and horizontal components that seems all edges are correlated. This is being used for face recognition that is made famous by technology companies.

</p>

```python
from scipy import signal
import scipy
```

## 4.1 Vertical

```python
pattern = [[-1, 2, -1], [-1, 2, -1], [-1, 2, -1]]
vip1 = rgb2gray(imread('C:\\Users\\MaryChrisGo\\Documents\\4th Year 2nd Sem AY 18-19\\Physics166\\vip.jpg'))
final = scipy.signal.convolve2d(pattern, vip1, mode='full', boundary='fill', fillvalue=0)
plt.imshow(final, cmap="gray")
plt.show()
```
