# Activity 1: Anamorphic Property of FT of different 2D patterns


<p style='text-align: justify;'>

FT modulus we obtained from the Fourier Transform of the images. Displayong anamorphism, we manage to check the central peak in the Fourier transform. For the second part, the distance between the dots affects the frequency of the generated images.  

</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as img
from skimage.color import rgb2gray
from skimage.io import imread
import matplotlib.image as mpimg
import cv2
%matplotlib inline
from numpy.fft import fft2, fftshift,fft,ifft2
```

## 1.1 Tall rectangle aperture

```python
a=img.imread('tall.png')
b=rgb2gray(a)
plt.imshow(b,cmap='gray')
plt.show()

FI = abs(np.fft.fft2(b))
FIS = np.fft.fftshift(FI)
plt.imshow(FIS, cmap = 'gray')
plt.show()
```

## 1.2 Wide rectangle aperture

```python
a=img.imread('wide.png')
b=rgb2gray(a)
plt.imshow(b,cmap='gray')
plt.show()

FI = abs(np.fft.fft2(b))
FIS = np.fft.fftshift(FI)
plt.imshow(FIS, cmap = 'gray')
plt.show()
```

## 1.3 Two dots along x-axis symmetric about the center


```python
a=img.imread('twodots.png')
b=rgb2gray(a)
plt.imshow(b,cmap='gray')
plt.show()

FI = abs(np.fft.fft2(b))
FIS = np.fft.fftshift(FI)
plt.imshow(FIS, cmap = 'gray')
plt.show()
```

## 1.4 Different spacing between dots

# Activity 2. Rotation Property of the FT

<p style='text-align: justify;'>

This simple shows the rotation property of FT. By increasing the frequency, we see more visible lines. Adding theta rotates the image.

</p>

```python
f = 4

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
sine = np.sin(2*np.pi*f*X)

plt.imshow(sine, cmap = 'gray')
plt.show()
sin1 = abs(np.fft.fft2(sine))
sin2 = np.fft.fftshift(sin1)
plt.imshow(sin2, cmap = 'gray')
plt.show()
```

```python
f = 100

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
sine = np.sin(2*np.pi*f*X)

plt.imshow(sine, cmap = 'gray')
plt.show()

sin1 = abs(np.fft.fft2(sine))
sin2 = np.fft.fftshift(sin1)
plt.imshow(sin2, cmap = 'gray')
plt.show()
```

```python
f = 4
theta = 30

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
z = np.sin(2*np.pi*f*(Y*np.sin(theta) + X*np.cos(theta)))

plt.imshow(z, cmap = 'gray')
plt.show()
sin1 = abs(np.fft.fft2(z))
sin2 = np.fft.fftshift(sin1)
plt.imshow(sin2, cmap = 'gray')
plt.show()
```

```python
f = 4
theta = 30

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
combination = (np.sin(2*np.pi*(f*X)))*(np.sin(2*np.pi*(f*Y)))

plt.imshow(combination, cmap = 'gray')
plt.show()
sin1 = abs(np.fft.fft2(combination))
sin2 = np.fft.fftshift(sin1)
plt.imshow(sin2, cmap = 'gray')
plt.show()

turned_combination = (np.sin(2*np.pi*(f*(X*np.cos(theta)))))*(np.sin(2*np.pi*(f*(Y*np.cos(theta)))))
plt.imshow(turned_combination, cmap = 'gray')
plt.show()
sin11 = abs(np.fft.fft2(turned_combination))
sin22 = np.fft.fftshift(sin11)
plt.imshow(sin22, cmap = 'gray')
plt.show()
```

# Activity 3:Convolution Theorem Redux


<p style='text-align: justify;'>

This first three parts of this activity deals with FT of the images. It can be observed from the final images. The twist here starts with the gaussian dots. Varying the variance, the pattern gets smaller. For the array of zeros, convolving A and d gives a random boxes placed inside the image. This means that the resulting images located the original Dirac deltas in the array. This shows that the integral of a Dirac delta being multipled to another function is the function itself. The next array of zeros showed a cross pattern. 


</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as img
from scipy.signal import convolve2d
%matplotlib inline
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

Image=img.imread('twodots.png')
gray=rgb2gray(Image)
A=gray
fti = abs(np.fft.fft2(A))
fsi = np.fft.fftshift(fti)
plt.imshow(fsi, cmap = 'gray')
```

```python
nx=100
x=np.linspace(-1,1,nx)
y=np.linspace(-1,1,nx)
X,Y=np.meshgrid(x,y)
sigma = 0.2
G = np.exp(-((X + 0.5)**2+Y**2)/sigma**2) + np.exp(-((X - 0.5)**2+Y**2)/sigma**2)
plt.imshow(G,cmap='gray')
plt.show()

FI = np.fft.fft2(G)
FIS = abs(np.fft.fftshift(FI))
plt.imshow(FIS,cmap='gray')
plt.show()
```

## Activity 4:Fingerprints : Ridge Enhancement

<p style='text-align: justify;'>

This activity showed on showing the image much bolder. From treating the FT of the image, we were able to enhance the ridges of the fingerprint.

</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as mpimg
from numpy.fft import fft2, fftshift, ifft2,ifftshift
```

```python
fingerprint = mpimg.imread('fingerprint.jpg',0)
fingerprint = fingerprint[:,:,0]
plt.imshow(fingerprint, cmap = 'gray')
plt.show()

f_shift =fftshift(fft2(fingerprint))
lnS = np.log(f_shift)
plt.imshow(abs(lnS),cmap='gray')
plt.imshow(abs(ifft2(fftshift(np.exp(lnS)))),cmap='gray')

x1 = np.linspace(-1,1,427)
y2 = np.linspace(-1,1,320)
X2,Y2 = np.meshgrid(y2,x1)
f1 = np.where((X2)**2 + (Y2)**2 < 0.05**2,1,0)
f2 = np.where((X2)**2 + (Y2)**2 < 0.45**2,1,0)
f3 = np.where((X2)**2 + (Y2)**2 < 0.10**2,1,0)
fn = np.add(f1,np.invert(f3))
plt.imshow(np.subtract(f1,f3),cmap='gray')
filt = np.add(np.subtract(f1,f3),f2)
m2 = np.multiply(filt,lnS)
plt.imshow(abs((ifft2(np.exp(m2)))),cmap='gray')
```

# Activity 5: Lunar Landing Scanned Pictures : Line removal


<p style='text-align: justify;'>

The filter was made using photoshop and was successful in removing the vertical lines from the image. Consider passing a light through the image. Think of the filter as the blocker of the light that passes through. As can be seen from the filter, there are white spots. Using photoshop, we "blocked" this. Multiply this to the Ft of the image and FT the final image gives you the desired image.
</p>

```python
from PIL import Image
moon=img.imread('hiresv.JPG')
gmoon=rgb2gray(moon)
plt.imshow(gmoon,cmap='gray')
plt.show()

f1 = abs(np.fft.fft2(gmoon))
f2 = np.log(abs(np.fft.fftshift(f1)))
plt.imshow(f2, cmap ='gray')

mask = img.imread('mask.jpg')
gmask = rgb2gray(mask)
shift_mask=np.fft.fftshift(gmask)

FI=np.fft.fft2(gmoon)
final = FI*shift_mask
IRA=np.fft.fft2(final)
FImage=abs(IRA)
plt.imshow(FImage,cmap='gray')
plt.show()

```
